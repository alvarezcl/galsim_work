```python
from __future__ import division
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import drawLibrary
import galsim
import lmfit
import matplotlib.pyplot as plt
import scipy.linalg as scla
import matplotlib.gridspec as gridspec
%matplotlib inline
pd.set_option('display.max_columns', None)
```

We begin with a simple covariance matrix that describes a 2D gaussian at an angle of 45 degrees

```python
cov = np.array([[1.58,0.5],[0.5,0.58]])
diag = np.diag(cov)
outer_diag = np.sqrt(np.outer(diag,diag))
correlation_mat = cov/outer_diag
mean = np.array([2,2])
x,y = np.random.multivariate_normal(mean,cov,8000).T
```

Now we diagonalize the covariance matrix.

```python
cov = DataFrame(cov, columns = ['x','y'],index = ['x','y'])
cov
```

```python
eval, evec = scla.eig(cov)
eval = DataFrame(np.real(np.diag(eval)))
eval
```

```python
evec = DataFrame(np.real(evec))
evec
```

Our original basis was in x,y and our new basis is:

```python
%%latex
$\mathbf{x_w} = \mathbf{W}^\intercal\mathbf{x}$
```

Where W is the matrix of eigenvectors.

```python
Wt = evec.T
Wt
```

```python
basis = np.array([[1,0],[0,1]])
X_w = DataFrame(Wt.dot(basis))
X_w
```

```python
Vectors = np.array([[mean[0],mean[0],X_w.values[0,0],X_w.values[0,1]],
                    [mean[0],mean[0],X_w.values[1,0],X_w.values[1,1]],
                    [0,0,1,0],
                    [0,0,0,1]])
Vectors
```

```python
X,Y,U,V = zip(*Vectors)
plt.figure()
ax = plt.gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1,zorder=10)
plt.draw()
plt.plot(x,y,'x',alpha=0.1)
plt.axis('equal')
plt.grid(True)
```

```python
correlation_mat = DataFrame(correlation_mat, columns = ['x','y'],index = ['x','y'])
correlation_mat
```

```python
eval_corr, evec_corr = scla.eig(correlation_mat)
eval_corr = DataFrame(np.real(np.diag(eval_corr)))
eval_corr
```

```python
evec_corr = DataFrame(np.real(evec_corr))
evec_corr
```

```python
Wt_corr = evec_corr.T
Wt_corr
```

```python
X_w_corr = DataFrame(Wt_corr.dot(basis))
X_w_corr
```

```python
Vectors_corr = np.array([[mean[0],mean[0],X_w_corr.values[0,0],X_w_corr.values[0,1]],
                         [mean[0],mean[0],X_w_corr.values[1,0],X_w_corr.values[1,1]],
                         [0,0,1,0],
                         [0,0,0,1]])
Vectors_corr
```

```python
X,Y,U,V = zip(*Vectors_corr)

ax = plt.gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1,zorder=10)
plt.draw()
plt.plot(x,y,'x',alpha=0.1)
plt.axis('equal')
plt.grid(True)
```
